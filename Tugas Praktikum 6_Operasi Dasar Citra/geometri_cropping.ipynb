```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# membaca citra
img = cv2.imread('lena.png')

# menentukan koordinat awal dan akhir untuk cropping
x1, y1 = 50, 50
x2, y2 = 250, 250

# melakukan cropping pada citra
crop_img = img[y1:y2, x1:x2]

# menampilkan citra asli dan hasil cropping
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB))
plt.title('Hasil Cropping')
plt.xticks([]), plt.yticks([])

plt.show()
```
