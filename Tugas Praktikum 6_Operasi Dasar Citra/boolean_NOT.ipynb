```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# membaca dua gambar yang akan dioperasikan
img1 = cv2.imread('cowo_ghibli.jpeg')
img2 = cv2.imread('rice.png')

# melakukan operasi boolean NOT pada kedua gambar
result1 = cv2.bitwise_not(img1)
result2 = cv2.bitwise_not(img2)

# menampilkan gambar sebelum dan sesudah operasi NOT
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Gambar 1 sebelum operasi NOT')

plt.subplot(2,2,2)
plt.imshow(cv2.cvtColor(result1, cv2.COLOR_BGR2RGB))
plt.title('Hasil operasi NOT pada gambar 1')

plt.subplot(2,2,3)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Gambar 2 sebelum operasi NOT')

plt.subplot(2,2,4)
plt.imshow(cv2.cvtColor(result2, cv2.COLOR_BGR2RGB))
plt.title('Hasil operasi NOT pada gambar 2')

plt.show()
```
