```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# membaca citra
img = cv2.imread('lena.png')

# melakukan flipping vertical
flipped_img = cv2.flip(img, 0)

# menampilkan citra asli
plt.subplot(121)
plt.imshow(img)
plt.title('Citra Original')

# menampilkan citra hasil flipping
plt.subplot(122)
plt.imshow(flipped_img)
plt.title('Citra Hasil Vertical Flipping')

# menampilkan plot
plt.show()
```
