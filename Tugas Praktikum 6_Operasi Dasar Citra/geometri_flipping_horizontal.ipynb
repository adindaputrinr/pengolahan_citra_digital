```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load gambar
img = cv2.imread('lena.png')

# Flip horizontal
flip_horizontal = cv2.flip(img, 1)

# Tampilkan gambar asli dan hasil flipping horizontal
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].set_title('Original')
ax[1].imshow(cv2.cvtColor(flip_horizontal, cv2.COLOR_BGR2RGB))
ax[1].set_title('Citra Hasil Horizontal Flipping')

plt.show()
```
