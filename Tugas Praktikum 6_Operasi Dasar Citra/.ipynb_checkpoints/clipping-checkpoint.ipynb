```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# membaca gambar
img = cv2.imread('cowo_ghibli.jpeg')

# mendefinisikan batas clipping
left = 100
right = 400
top = 100
bottom = 300

# melakukan clipping pada gambar
clipped_img = img[top:bottom, left:right]

# menampilkan gambar sebelum dan sesudah clipping
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Gambar sebelum clipping')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(clipped_img, cv2.COLOR_BGR2RGB))
plt.title('Gambar setelah clipping')

plt.show()
```
