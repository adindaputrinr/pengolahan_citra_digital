```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Baca gambar
img = cv2.imread('cowo_ghibli.jpeg')

# Tentukan koordinat x dan y yang akan digunakan untuk translasi
x = 50
y = 100

# Tentukan matriks translasi
M = np.float32([[1, 0, x], [0, 1, y]])

# Lakukan operasi translasi menggunakan fungsi warpAffine
translated_img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

# Tampilkan gambar sebelum dan sesudah operasi translasi
fig, axes = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = axes.ravel()

ax1.imshow(img)
ax1.set_title('Original')

ax2.imshow(translated_img)
ax2.set_title('Hasil')

plt.show()
```
