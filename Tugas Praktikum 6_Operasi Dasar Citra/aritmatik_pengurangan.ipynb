```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# membaca dua buah citra/gambar
f = cv2.imread('cameraman.tif', 0)
g = cv2.imread('rice.png', 0)

# mengambil ukuran citra/gambar
r1, c1 = f.shape
r2, c2 = g.shape

# melakukan operasi pengurangan pada setiap piksel
if (r1 == r2) and (c1 == c2):
    h = np.zeros((r1, c1), dtype=np.uint8)
    for x in range(r1):
        for y in range(c1):
            h[x, y] = f[x, y] - g[x, y]

# membatasi nilai piksel antara 0 dan 255
h = np.clip(h, 0, 255)

# menampilkan citra/gambar asli dan citra/gambar hasil pengurangan
plt.figure(figsize=(10, 5))
plt.subplot(131)
plt.imshow(f, cmap='gray')
plt.title('Citra 1')
plt.subplot(132)
plt.imshow(g, cmap='gray')
plt.title('Citra 2')
plt.subplot(133)
plt.imshow(h, cmap='gray')
plt.title('Citra Gabungan')
plt.show()
```
