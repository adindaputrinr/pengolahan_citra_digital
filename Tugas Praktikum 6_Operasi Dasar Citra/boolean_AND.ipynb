```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# membaca dua gambar yang akan dioperasikan
img1 = cv2.imread('cowo_ghibli.jpeg')
img2 = cv2.imread('rice.png')

# menyesuaikan ukuran gambar agar sama
img1 = cv2.resize(img1, (img2.shape[1], img2.shape[0]))

# melakukan operasi boolean AND pada kedua gambar
result = cv2.bitwise_and(img1, img2)

# menampilkan gambar sebelum dan sesudah operasi AND
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Gambar 1')

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Gambar 2')

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title('Hasil operasi AND')

plt.show()
```
