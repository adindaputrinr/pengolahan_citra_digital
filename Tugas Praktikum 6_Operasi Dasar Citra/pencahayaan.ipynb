```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# membaca gambar
img = cv2.imread('cowo_ghibli.jpeg')

# mendefinisikan scalar untuk pencahayaan
alpha = 2.0

# melakukan operasi pencahayaan pada gambar
bright_img = cv2.addWeighted(img, alpha, np.zeros(img.shape, img.dtype), 0, 0)

# menampilkan gambar sebelum dan sesudah pencahayaan
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(bright_img, cv2.COLOR_BGR2RGB))
plt.title('Pencahayaan')

plt.show()
```
